#### Training ML model

```python
stats.columns
```

```python
# to  make a prediction we need al the numerical columns. aslo we need to get rid of the columns that are directly correlated with Share
# since it is what we are trying to predict. If we include them it will cause the algorithm to overfit
predictors = ["Age", "G", "GS", "MP", "FG", "FGA", 'FG%', '3P', '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%', 'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS', 'W', 'L', 'W/L%',
       'GB', 'PS/G', 'PA/G', 'SRS']
```

```python
train= stats[stats["Year"]<2021]
test = stats[stats["Year"]==2021]
```

```python
# we will use ridge regression. It is similar to lineer regression but it prevents overfitting by shrinking the lineeer regression 
# coefficients.
from sklearn.linear_model import Ridge

reg = Ridge(alpha=.1)
```

```python
reg.fit(train[predictors], train["Share"])
```

```python
predictions=reg.predict(test[predictors])
predictions = pd.DataFrame(predictions, columns=["predictions"], index=test.index)
```

```python
predictions
```

```python
combination= pd.concat([test[["Player","Share"]],predictions],axis=1)
```

```python
combination.sort_values("Share",ascending=False).head(10)
```

```python
# define an error metric
#calculate how many players we placed correctly to top 5
# question is if we placed the person that should actaully be in the top 5, into the top 5 in predicted ranks.
# go down the rows acooording to predicted ranks and add a point when you see a player that is in the actual top 5
def find_ap(combination):
    actual = combination.sort_values("Share", ascending=False).head(5)
    predicted = combination.sort_values("predictions", ascending=False)
    ps = []
    found = 0
    seen = 1
    for index,row in predicted.iterrows():
        if row["Player"] in actual["Player"].values:
            found += 1
            ps.append(found / seen)
        seen += 1
    #print("ps:", ps)
    return sum(ps) / len(ps)
```

```python
find_ap(combination) # the higher the value the better
```

#### Back testing

```python
years=list(range(1991,2022))
```

```python
aps = []
all_predictions = []
for year in years[5:]:
    train = stats[stats["Year"] < year]
    test = stats[stats["Year"] == year]
    reg.fit(train[predictors],train["Share"])
    predictions = reg.predict(test[predictors])
    predictions = pd.DataFrame(predictions, columns=["predictions"], index=test.index)
    combination = pd.concat([test[["Player", "Share"]], predictions], axis=1)
    all_predictions.append(combination)
    aps.append(find_ap(combination))
```

```python
def add_ranks(combination):
    combination = combination.sort_values("predictions", ascending=False)
    combination["Predicted_Rk"] = list(range(1,combination.shape[0]+1))
    combination = combination.sort_values("Share", ascending=False)
    combination["Rk"] = list(range(1,combination.shape[0]+1))
    combination["Diff"] = (combination["Rk"] - combination["Predicted_Rk"])
    return combination
```

```python
add_ranks(all_predictions[1])
```

```python
rank=add_ranks(all_predictions[1])
rank.sort_values(["Rk"],ascending=True).head(5)
```

```python
def backtest(stats, model, years, predictors):
    aps = []
    all_predictions = []
    for year in years:
        train = stats[stats["Year"] < year]
        test = stats[stats["Year"] == year]
        model.fit(train[predictors],train["Share"])
        predictions = model.predict(test[predictors])
        predictions = pd.DataFrame(predictions, columns=["predictions"], index=test.index)
        combination = pd.concat([test[["Player", "Share"]], predictions], axis=1)
        combination = add_ranks(combination)
        all_predictions.append(combination)
        aps.append(find_ap(combination))
    return sum(aps) / len(aps), aps, pd.concat(all_predictions)
```

```python
mean_ap, aps, all_predictions = backtest(stats, reg, years[5:], predictors)
```
