#### Merge mvps and players data sets on player and year column
Before the merge we need to get rid of the * that is in the player names in players df. Because of this * the names in the mvps and player are not the same. And this doesn't allow us to merge these two dfs on the Player column.
Also in players df some names show up more than once but we need each name to show up once for the merge operation.

```python
players["Player"].head(50)
```

```python
players["Player"]=players["Player"].str.replace("*","",regex=False)
```

```python
players.head(50)
```

```python
players.groupby(["Player","Year"])
```

```python
players.groupby(["Player","Year"]).get_group(("Greg Anderson",1991))
```

```python
# each group need to be made of one row. In order to do this we write a single_row function 
#and it will take each group as an argument
def single_team(df):
    if df.shape[0]==1:
        return df
    else:
        row = df[df["Tm"]=="TOT"]
        row["Tm"] = df.iloc[-1,:]["Tm"]
        return row

players = players.groupby(["Player", "Year"]).apply(single_team)
```

```python
players.index = players.index.droplevel()
players.index = players.index.droplevel()
```

```python
players[players["Player"]=="Greg Anderson"]
```

```python
players.head()
```

```python
combined= players.merge(mvps,how= "outer",on=["Player", "Year"])#how= "outer"=> means even if the player isn't in the mvp table keep 
                                                                #the data in the combined table
```

```python
combined.head()
```

```python
combined[combined["Pts Won"]>0]
```

```python
#replace the Nan values in the tsble with 0's
combined[["Pts Won","Pts Max","Share"]]=combined[["Pts Won","Pts Max","Share"]].fillna(0)
```

```python
combined
```

#### Cleaning the teams data

```python
teams = pd.read_csv("teams.csv")
```

```python
teams.head()
```

```python
teams["Team"]=teams["Team"].str.replace("*","",regex=False)
```

```python
teams.head(2)
```

```python
teams["Team"].unique() # prints the unique team names in the data set
```

```python
combined["Tm"].unique()
```

```python
# Matcking the teams names with their shortend versions
nicknames={}
with open("nicknames.csv") as f:
    lines=f.readlines()
for line in lines[1:]:
    abbrev, name=line.replace("\n","").split(",")
    
    nicknames[abbrev]=name
```

```python
nicknames
```

```python
combined["Team"] = combined["Tm"].map(nicknames)
#pandas.map(), which is used to map values from two series having one similar column. For mapping two series, the last column of 
# the first should be the same as the index column of the second series, also the values should be unique.
```

```python
combined.head()
```

```python
stats=combined.merge(teams,how="outer",on=["Team","Year"])
```

```python
stats
```

```python
stats.drop(["Unnamed: 0"],axis=1, inplace=True)
```

```python
stats.dtypes # need to chech the data types of the columns before using them
```

```python
stats = stats.apply(pd.to_numeric, errors='ignore') #converts the string values that can be numbers into numerical values 
```

```python
stats.dtypes
```

```python
stats["GB"] = pd.to_numeric(stats["GB"].str.replace("—","0"))
```

```python
stats.dtypes
```

```python
stats.to_csv("player_mvp_stats.csv")
```
